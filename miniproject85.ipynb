## Sales Data Analysis

/// Objective: Analyze a sales dataset to find the top-selling products and the
average revenue per region.//

```python
import pandas as pd
```

///Import data:

```python
data = {
    'product': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B'],
    'region': ['North', 'South', 'East', 'West', 'North', 'East', 'South', 'West'],
    'units_sold': [10, 15, 8, 12, 20, 5, 18, 10],
    'price': [100, 150, 100, 200, 150, 100, 200, 150]
}
df=pd.DataFrame(data)
```

///find top selling Product//

```python
df['total_sales']=df['units_sold']*df['price']
top_product=df.groupby('product')["total_sales"].sum().sort_values(ascending=False).reset_index()

print("Total sales of each product is:\n",top_product)
```

///Average revenue per region///

```python
avg_revenue=df.groupby('region')['total_sales'].mean().reset_index()
print("Average revenue per region is:\n",avg_revenue)
```
