## Movie Data Cleaning

/////Objective: Clean a movie dataset that contains duplicate entries and
inconsistent values for a genre column.//

```python
import pandas as pd
```

///import data//

```python
data = {
    'title': ['Inception', 'Avatar', 'Inception', 'Titanic', 'avatar'],
    'genre': ['Sci-Fi', 'sci fi', 'Sci-Fi', 'Romance ', 'romance']
}
df=pd.DataFrame(data)
```

//Remove Duplicate Movie enteries//

```python
df_cleaned = df.drop_duplicates(subset=['title', 'genre'], keep='first').copy()
print(df_cleaned)
```

//clean and standardize genre column

```python
df_cleaned['genre'] = (
    df_cleaned['genre']
        .str.lower()            # convert to lowercase
        .str.strip()            # remove extra spaces
        .str.replace('-', ' ')  # replace hyphens with spaces
        .str.replace('_', ' ')  # replace underscores
)
print("Cleaned movie dataset:\n",df_cleaned)
```
